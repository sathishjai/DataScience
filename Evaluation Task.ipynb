```python
# Outlier Detection
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=df["Price"])
```

```python
def remove_outliers_iqr(df, feature):
    Q1=df[feature].quantile(0.25)
    Q3=df[feature].quantile(0.75)
    IQR=Q3-Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df [(df [feature] >= lower_bound) & (df[feature] <= upper_bound)]

A = remove_outliers_iqr(df,'Price')
print("Before Cleaning", df.shape)
print("After Cleaning", A.shape)
```

##### Converting the Columns Mileage, Engine, Power to numerical values in DF

```python
df_copy = df.copy()
```

```python
# Converting the Columns Mileage,Engine,Power to numerical values
import re
# Function to extract numeric values
def extract_numeric_value(text):
    numeric_value = re.findall(r"[-+]?\d*\.\d+|\d+", text)
    return float(numeric_value[0]) if numeric_value else None

# Apply the function to the DataFrame column
df_copy['Mileage_in_kmpl'] = df_copy['Mileage'].apply(extract_numeric_value)
df_copy['Engine_in_CC'] = df_copy['Engine'].apply(extract_numeric_value)
df_copy['Power_in_bhp'] = df_copy['Power'].apply(extract_numeric_value)
df_copy.head()
```

```python
columns_to_remove = ['Mileage', 'Engine', 'Power']
df_copy = df_copy.drop(columns=columns_to_remove)
```

```python
# Numerical Features
from sklearn.feature_selection import SelectKBest, f_classif
x = df_copy[['Year','Kilometers_Driven','Seats','Mileage_in_kmpl','Engine_in_CC','Power_in_bhp']] # Numerical Features
y = df_copy['Price']

f_p_values=f_classif(x,y)
f_p_values
```

```python
f_p_values=pd.Series(f_p_values[1])
f_p_values.index=x.columns
f_p_values
```

##### Features (Auto)

```python
best_features = SelectKBest(score_func=f_classif, k=2)
fit = best_features.fit(x,y)
selected_features = x.columns[fit.get_support()]
```

```python
selected_features
```

```python
sns.pairplot(df_copy,hue='Price')
```

```python
# Categorical Features
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Identify categorical features
categorical_features = df_copy.select_dtypes(include=['object']).columns

anova_results = {}
for feature in categorical_features:
    model = ols(f'Price ~ C({feature})', data=df_copy).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    anova_results[feature] = anova_table['PR(>F)'][0]

# Sort features by p-value
sorted_anova_results = sorted(anova_results.items(), key=lambda x: x[1])

# Display the sorted ANOVA results
print("ANOVA Results (sorted by p-value):")
for feature, p_value in sorted_anova_results:
    print(f"Feature: {feature}, p-value: {p_value}")
```
