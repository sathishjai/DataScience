```python
#Task -8  Chi- test & Anova
# Reconnect to the database and load the data
data2 = connect(host='localhost', user='root', passwd='Manju_004', database='Ecommerce_sales')
query = 'select * from Customer_Order_Product_Details'
df = pd.read_sql(query, data2)
df
```

```python
df['category'].unique()
```

```python
df['quantity'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
lable=LabelEncoder()

# Encoding the categorical variables:
df['category']=lable.fit_transform(df['category'])
df['quantity']=lable.fit_transform(df['quantity'])
df['total_price'] = lable.fit_transform(df['total_price'])
df['stock_quantity'] = lable.fit_transform(df['stock_quantity'])
df['price'] = lable.fit_transform(df['price'])

# Selecting specific columns:
selected_columns = ['category','stock_quantity', 'quantity','price','total_price']
df_selected = df[selected_columns]

# Printing the selected DataFrame:
print(df_selected)
```

```python
# chi2 test 
from sklearn.feature_selection import chi2, SelectKBest

# Performing chi2 test:
chi2_values, p_values_chi2 = chi2(df_selected[['category', 'quantity']], df_selected['total_price'])

# creating a dataframe for the results:
chi2_results = pd.DataFrame({"Feature": ['category', 'quantity'], "Chi2 Stat": chi2_values, "p-value": p_values_chi2})

# printing the results:
print("Chi2 Test Results:")
print(chi2_results)
```

```python
df_selected.columns
```

```python
# One-way ANOVA
from sklearn.feature_selection import f_classif

# Performing the one-way ANOVA test: 
f_values, p_values_f = f_classif(df_selected[['total_price','price','stock_quantity']], df_selected['category'])

#creating a dataframe for the results:
f_test_results = pd.DataFrame({"Feature": ['total_price','price','stock_quantity'], "F Stat": f_values, "p-value": p_values_f})

# printing the results:
print("ANOVA Test Results:")
print(f_test_results)
```

```python
# Select the two columns with the highest dependency using Chi2 and F-test
chi2_selected_features = chi2_results[chi2_results['p-value'] < 0.05]['Feature'].values
f_test_selected_features = f_test_results[f_test_results['p-value'] < 0.05]['Feature'].values

# print the results:
print("\nSelected Features based on Chi2 Test:", chi2_selected_features)
print("Selected Features based on F-test:", f_test_selected_features)
```
