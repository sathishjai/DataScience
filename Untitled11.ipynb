```python
df = df.dropna(subset=['class','alone', 'alive'])
```

```python
df = df.dropna(subset=['sex','class', 'alone', 'alive','who'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['alone'] = label_encoder.fit_transform(df['alone'])
df['who'] = label_encoder.fit_transform(df['who'])
df['alive'] = label_encoder.fit_transform(df['alive'])
df['class'] = label_encoder.fit_transform(df['class'])
df
```

```python
df.head()
```

```python
df['class'].unique()
```

```python
intp= df[['sex','class', 'alone','who']]
out= df['alive']
```

```python
from sklearn.feature_selection import chi2
```

```python
chi2_p,pvalue = chi2(intp,out)
chi2_p,pvalue
```

```python
chi2_df = pd.DataFrame({'intp': intp.columns,'Chi2': chi2,'Pvalue': pvalue})
chi2_df
```
