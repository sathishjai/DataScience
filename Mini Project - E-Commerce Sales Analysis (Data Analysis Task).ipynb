### Let me know which two columns are highly dependent to each other using Chi2 and F_classif(ANOVA)

```python
# Using Chi-Square Test :
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
label_encoders = {}
dfcopy = df.copy()
categorical_columns = ['customer_name', 'email', 'join_date', 'order_date', 'product_name', 'category']

for i in categorical_columns:
    dfcopy[i] = label.fit_transform(dfcopy[i])  # Changed 'encoded_df' to 'dfcopy'
    label_encoders[i] = label
    
```

```python
# Splitting Categorical and Numerical Data from the Data for Chi-Square Test :
categorical_data = dfcopy[categorical_columns]
numerical_data = dfcopy[['price', 'stock_quantity', 'quantity', 'total_price']]
```

### >> Chi-Square Test

```python
from sklearn.feature_selection import chi2
target = dfcopy['quantity'] # Using 'quantity' as target for chi2
f_p_values=chi2(categorical_data,target)
```

```python
f_p_values
```

```python
p_values=pd.Series(f_p_values[1])
p_values.index=categorical_data.columns
print('P_values with quantity:')
p_values
```

```python
# For Checking Dependancy on among Categorical Data :
from sklearn.feature_selection import chi2
x = 1
for _ in range(6):
    if x == 1:
        target = dfcopy[["category"]]
        target_name = "category"
    elif x == 2:
        target = dfcopy[["product_name"]]
        target_name = "product_name"
    elif x == 3:
        target = dfcopy[["order_date"]]
        target_name = "order_date"
    elif x == 4:
        target = dfcopy[["join_date"]]
        target_name = "join_date"
    elif x == 5:
        target = dfcopy[["email"]]
        target_name = "email"
    elif x == 6:
        target = dfcopy[["customer_name"]]
        target_name = "customer_name"
    
    f_p_values = chi2(dfcopy[categorical_columns], target)
    p_values = pd.Series(f_p_values[1])
    p_values.index = categorical_columns
    print(f"P-values for {target_name} (x={x}):")
    print(p_values)
    print()
    x += 1
```
