## Facebook REVIEW Dataset (from Kaggle) :

```python
import pandas as pd
from sklearn.datasets import fetch_openml

# Load the Ames housing dataset
ames_housing = fetch_openml(name="house_prices", as_frame=True)

# Display the shape of the data
print(ames_housing.data.shape)

# Display the feature names
print(ames_housing.feature_names)

# Convert to DataFrame for easier handling
ames_df = ames_housing.frame
print(ames_df.head())
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
boston_c = ames_df
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(boston_c['LotArea'], boston_c['SalePrice' ])
plt.show()
```
