#### Missing Value Imputation :

```python
df['UserName'].unique()
```

```python
df_mode = df.UserName.mode()[0]
```

```python
df.UserName.fillna(df_mode,inplace=True)
```

```python
df.isnull().sum()
```

#### Content

```python
df['Content'].unique()
```

```python
df_mode2 = df.Content.mode()[0]
```

```python
print(df_mode2)
```

```python
df.Content.fillna(df_mode2,inplace=True)
```

```python
df.isnull().sum()
```
